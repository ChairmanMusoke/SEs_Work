```python
import streamlit as st
import pandas as pd
import os

# Load data
@st.cache
def load_data("data/Village_ScoresA.csv"):
    return pd.read_csv("data/Village_ScoresA.csv")

# Function to display summaries
def display_summary(filtered_data, variables):
    st.write("### Summary")
    for var in variables:
        if var in filtered_data.columns:
            mean_val = filtered_data[var].mean()
            st.write(f"**{var}**: {mean_val:.2f}%")
        else:
            st.write(f"**{var}**: Not found in data")

def main():
    st.title("Village Level Data Summary App")

    # Define the path to the data file
    data_file_path = os.path.join('data', 'your_data_file.csv')  # Replace 'your_data_file.csv' with your actual file name
    
    # Load the data
    data = load_data(data_file_path)
    
    # Filter widgets
    district = st.selectbox("Select District", sorted(data['district_0'].unique()))
    cluster = st.selectbox("Select Cluster", sorted(data[data['district_0'] == district]['cluster_0'].unique()))
    village = st.selectbox("Select Village", sorted(data[(data['district_0'] == district) & (data['cluster_0'] == cluster)]['village_0'].unique()))
    
    # Filter data based on selections
    filtered_data = data[(data['district_0'] == district) & (data['cluster_0'] == cluster) & (data['village_0'] == village)]
    
    # Select variables for summary
    all_columns = [col for col in data.columns if col not in ['district_0', 'cluster_0', 'village_0']]
    selected_vars = st.multiselect("Select variables to summarize", all_columns, default=all_columns)

    # Display summary
    display_summary(filtered_data, selected_vars)

if __name__ == "__main__":
    main()
```
