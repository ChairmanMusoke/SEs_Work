```python
import pandas as pd
import numpy as np
```

```python
BackCheck = pd.read_csv("/Users/rtv-lpt-129/Downloads/RTV Remote Survey - 3.csv", encoding='ISO-8859-1', low_memory=False)
```

```python
Data = BackCheck[(BackCheck['status'] == 1) & (BackCheck['consent_1'] == 1)].copy()
```

```python
Data['season1crops'] = Data['Season_1'].apply(lambda x: len(x.split()) if isinstance(x, str) and x != '99' else 0)
Data['season2crops'] = Data['Season_2'].apply(lambda x: len(x.split()) if isinstance(x, str) and x != '99' else 0)
Data['AHSseason1crops'] = Data['ahs_Season_1'].apply(lambda x: len(x.split()) if isinstance(x, str) and x != '99' else 0)
Data['AHSseason2crops'] = Data['ahs_Season_2'].apply(lambda x: len(x.split()) if isinstance(x, str) and x != '99' else 0)
Data['Perennials'] = Data['perennial_crops_grown'].apply(lambda x: len(x.split()) if isinstance(x, str) and x != '99' else 0)
Data['AHSPerennials'] = Data['ahs_perennial_crops_grown'].apply(lambda x: len(x.split()) if isinstance(x, str) and x != '99' else 0)
Data['LivestockNum'] = Data['Livestock_list'].apply(lambda x: len(x.split()) if isinstance(x, str) and x != '99' else 0)
Data['AHSLivestockNum'] = Data['ahs_Livestock_list'].apply(lambda x: len(x.split()) if isinstance(x, str) and x != '99' else 0)
```

```python
Data['Age_Diff'] = Data['hhh_age'] - Data['ahs_age']
Data['Season1_Diff'] = Data['season1crops'] - Data['AHSseason1crops']
Data['Season2_Diff'] = Data['season2crops'] - Data['AHSseason2crops']
Data['Perennial_Diff'] = Data['Perennials'] - Data['AHSPerennials']
Data['Livestock_Diff'] = Data['LivestockNum'] - Data['AHSLivestockNum']
```

```python
Data['OwnShoes'] = np.where((Data['ahs_Every_Member_at_least_ONE_Pair_of_Shoes'] == 0) & (Data['Every_Member_at_least_ONE_Pair_of_Shoes'] == 1),0,1)
Data['SleepUnderMosquitoNet'] = np.where((Data['ahs_hh_member_sleep_under_net'] == 0) & (Data['hh_member_sleep_under_net'] == 1),0,1)
Data['Ownland'] = np.where((Data['ahs_Does_your_Household_own_any_Land'] == 0) & (Data['Does_your_Household_own_any_Land'] == 1),0,1)
Data['Boughtland'] = np.where((Data['ahs_land_bought'] == 0) & (Data['land_bought'] == 1),0,1)
Data['Season1Cropped'] = np.where((Data['ahs_Season1_cropped'] == 0) & (Data['Season1_cropped'] == 1),0,1)
Data['Season2Cropped'] = np.where((Data['ahs_Season2_cropped'] == 0) & (Data['Season2_cropped'] == 1),0,1)
Data['Improvedseeds'] = np.where((Data['ahs_Improved_seed'] == 0) & (Data['Improved_seed'] == 1),0,1)
Data['PerennialCropping'] = np.where((Data['ahs_perennial_cropping'] == 0) & (Data['perennial_cropping'] == 1),0,1)
Data['CasualLabour'] = np.where((Data['ahs_work_casual'] == 0) & (Data['work_casual'] == 1),0,1)
Data['LivestockPresent'] = np.where((Data['ahs_livestock_present_hh'] == 0) & (Data['livestock_present_hh'] == 1),0,1)
Data['VSLAParticipation'] = np.where((Data['ahs_vsla_Participation_Group'] == 0) & (Data['vsla_Participation_Group'] == 1),0,1)
Data['Borrowed12months'] = np.where((Data['ahs_borrowed_past_12_months'] == 0) & (Data['borrowed_past_12_months'] == 1),0,1)
Data['AnimalProduce'] = np.where((Data['ahs_Dairy_product_produce'] == 0) & (Data['Dairy_product_produce'] == 1),0,1)
```

```python
BackchecksOutput = Data[['starttime', 'sid_2', 'pre_district', 'pre_parish', 'pre_village', 'pre_name', 'ahs_surveyor_name', 
                         'hhh_survey', 'date_survey', 'survey_time', 'ahs_age', 'hhh_age', 'Age_Diff', 'season1crops', 'AHSseason1crops',
                         'Season1_Diff', 'season2crops', 'AHSseason2crops', 'Season2_Diff', 'Perennials', 'AHSPerennials', 'Perennial_Diff',
                         'LivestockNum', 'AHSLivestockNum', 'Livestock_Diff', 'OwnShoes', 'SleepUnderMosquitoNet',  'Ownland', 'Boughtland', 
                         'Season1Cropped', 'Season2Cropped', 'Improvedseeds', 'PerennialCropping', 'CasualLabour', 'LivestockPresent', 
                         'VSLAParticipation', 'Borrowed12months', 'AnimalProduce']]
```
